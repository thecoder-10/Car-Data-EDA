### DATA CLEANING

The company name is combine with the car model
<br>First, we will seperate the company and model into 2 different columns

```python
#Splitting company name from CarName column
new = car["CarName"].str.split(" ", n = 1, expand = True)
  
# making separate first name column from new data frame
car["Company"]= new[0]
  
# making separate last name column from new data frame
car["Model"]= new[1]
  
# Dropping old Name columns
car.drop(columns =["CarName"], inplace = True)
```

```python
car.head()
```

Two new columns, Company and Model have been successfully added

```python
#unique values in Company column

car['Company'] = car['Company'].str.capitalize()
car['Company'].unique()
```

Description of Company columns shows multiple names for a single company
<br> Changing incosistent company names
<br><br>
<b>   Correct Name :: Incosistent Name  </b>
<br> |     mazda    ::       maxda      
<br> |    porsche   ::       porcshce    
<br> |    toyota    ::      toyouta     
<br> |  volkswagon  ::  vokswagen, vw   


```python
#Updating values

car.Company.replace('Maxda','Mazda',inplace=True)
car.Company.replace('Porcshce','Porsche',inplace=True)
car.Company.replace('Toyouta','Toyota',inplace=True)
car.Company.replace('Vokswagen','Volkswagen',inplace=True)
car.Company.replace('Vw','Volkswagen',inplace=True)
```

```python
car['Company'].unique() #checking for updated values
```

#### Plotting Scatter Plot 

```python
#plotting imp_features

def plot_count(x, fig):   #histogram function for visualization
    plt.subplot(5,2,fig)
    plt.scatter(car[x], car['price'])
    plt.title(x+' vs price')
    plt.ylabel("Price")
    plt.xlabel(x)

plt.figure(figsize=(14,20))


plot_count("carwidth",1)
plot_count("curbweight",2)
plot_count("enginesize",3)
plot_count("horsepower",4)
plot_count("citympg",5)
plot_count("highwaympg",6)

plt.tight_layout()
```

<b> Inference :</b><br>
1) There is roughly linear positive relationship with price for curbweight, enginesize, and horsepower. In most cases, when these variables increase, the price increases too.
<br>2) There is an almost linear negative relationship for price with City's MPG and Highway's MPG. Lower MPG means more efficiency, hence the higher price.

### EDA on Categorical Data

```python
#working on categorical data

cat_data
```

```python
#Plotting a frequency table for cat_data

def frequency(x):   #helper function
    sns.barplot(x=car[x].value_counts().index, y=car[x].value_counts())
    plt.xlabel(x)
    plt.ylabel("Frequency")
    plt.title("Frequency table for %s" %x)
    plt.show()
    
for features in cat_data:
    frequency(features)
```

<b>Inference :</b>
<br>1)Gas is a more preferable fueltype than Diesel
<br>2)The count of vehicles that have standard aspiration is very high compared to the vehicles that have turbo aspiration.
<br>3)Both types of doornumber are almost equally preferable, four being a bit higher
<br>4)Sedans are most popular while Convertable cars are least used
<br>5)Almost all cars have Front Engine.
<br>6)Ohc is the most popular engine type and dohcv is least used
<br>7)Maximum cars have four cylinder number
<br>8)MPFI fuel system has the highest count whereas SPFI fuel system has the lowest count.

```python
#frequency table for company

plt.figure(figsize=(20,6))

frequency('Company')
```

<b>Inference :</b> Toyota is the most popular car company, while Mercury is least preferable by buyers
